## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straightforward, creating multiple series and putting them all together at the end.
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, \
# and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function.


regimen_group_stats = clean_merged_data.groupby('Drug Regimen').agg\
                        (
                        tumor_vol_Mean = ('Tumor Volume (mm3)', np.mean),
                        tumor_vol_Median = ('Tumor Volume (mm3)', np.median),
                        tumor_vol_Variance = ('Tumor Volume (mm3)', np.var),
                        tumor_vol_StDeviation = ('Tumor Volume (mm3)', np.std),
                        tumor_vol_SEM = ('Tumor Volume (mm3)', st.sem)                                                                    
                        ).round(2)


regimen_group_stats
```

## Bar Plots

```python
group = clean_merged_data.groupby(['Drug Regimen', 'Timepoint', 'Mouse ID'])

group_count = pd.DataFrame(group['Mouse ID'].count())
group_count.head(40)
```

```python
# Create a dataframe grouping by the 'Drug Regimen' column
regimen_group = clean_merged_data.groupby('Drug Regimen')

regimen_group_count = pd.DataFrame(regimen_group['Drug Regimen'].count())
```

```python
# Generate a bar plot showing the number of mice per time point \
# for each treatment throughout the course of the study using pandas.


count_chart = regimen_group_count.plot(kind='bar', title="Count of data points for each Drug Regimen", color="turquoise")

count_chart.set_xlabel("Drug Regimen")
count_chart.set_ylabel("Count")

plt.tight_layout()

plt.show()
```

```python
# Generate a bar plot showing the number of mice per time point \
# for each treatment throughout the course of the study using pyplot.

x_axis = np.arange(len(regimen_group_count))
tick_locations = [value for value in x_axis]


plt.figure(figsize=(7,3.5))
plt.bar(x_axis, regimen_group_count["Drug Regimen"], color='turquoise', align="center", width = 0.5)

plt.xticks(tick_locations, list(regimen_group_count.index), rotation="vertical")


# Set your x and y limits
plt.xlim(-0.65, len(x_axis)-0.35)
plt.ylim(0, max(regimen_group_count["Drug Regimen"])+10)


# Add axis labels and chart title
plt.title("Count of data points for each Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Count")

column_name = ["Drug Regimen"]
plt.legend(column_name,loc="best")


plt.show()
```

## Pie Plots

```python
# Create a dataframe grouping by the 'Sex' column
gender_group = clean_merged_data.groupby('Sex')


gender_group_count = pd.DataFrame(gender_group['Sex'].count())

# View the groupby dataframe
gender_group_count.head()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas

# Generate a pie plot showing the distribution of female versus male mice using pandas

gender_group_count.plot(kind='pie', y='Sex', \
                        title="Distribution of female versus male mice", \
                        colors=["pink", "skyblue"], \
                        shadow=True, \
                        startangle=120, \
                        autopct='%1.1f%%', \
                        legend =False)

plt.axis("equal")

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot

labels = list(gender_group_count.index.values)

piechart_val = gender_group_count['Sex']

colors =["pink", "skyblue"]


plt.pie(piechart_val, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=120)

plt.axis("equal")
plt.title("Distribution of female versus male mice")
plt.ylabel("Sex")


plt.show()
```
