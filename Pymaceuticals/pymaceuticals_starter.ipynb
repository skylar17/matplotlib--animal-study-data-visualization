```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np


# Hide warning messages in jupyter notebook(found this on Stack Overflow)
import warnings
warnings.filterwarnings('ignore')

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

```

```python
# View the dataframe
mouse_metadata.head()
```

```python
# Display the dataframe
study_results.head()
```

```python
# Combine the data into a single dataset
merged_data = pd.merge(mouse_metadata, study_results, on='Mouse ID')

# Display the merged DF
merged_data.head()
```

```python
# Checking the number of mice in the DataFrame.
merged_data['Mouse ID'].nunique()
```

```python
# Checking the number of times each mouse appeared in the DataFrame.
merged_data['Mouse ID'].value_counts()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
timepoint_group = merged_data.groupby(['Mouse ID','Timepoint'])

timepoint_group.sum().head(30) 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_id = merged_data.set_index('Mouse ID')
mouse = mouse_id.loc['g989', :]
mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# mouse_id = merged_data.set_index('Mouse ID')
drop_dupli_mouse = mouse_id.drop('g989')

clean_merged_data = drop_dupli_mouse.reset_index()
clean_merged_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_merged_data['Mouse ID'].value_counts()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straightforward, creating multiple series and putting them all together at the end.
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, \
# and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function.


regimen_group_stats = clean_merged_data.groupby('Drug Regimen').agg\
                        (
                        tumor_vol_Mean = ('Tumor Volume (mm3)', np.mean),
                        tumor_vol_Median = ('Tumor Volume (mm3)', np.median),
                        tumor_vol_Variance = ('Tumor Volume (mm3)', np.var),
                        tumor_vol_StDeviation = ('Tumor Volume (mm3)', np.std),
                        tumor_vol_SEM = ('Tumor Volume (mm3)', st.sem)                                                                    
                        ).round(2)


regimen_group_stats
```
